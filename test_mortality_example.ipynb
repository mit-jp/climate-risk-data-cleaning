```python
import pandas
import seaborn
import numpy
seaborn.set_theme(style="whitegrid")
```

# Estimated data

In this example, the estimated data falls far outside the distribution of the true data, indicating it may not be based on their mean.

```python
estimated_data = pandas.read_csv("Parsed data/Respiratory Mortality with estimates.csv")
seaborn.histplot(estimated_data["Percent Deaths 25+"])
```

# Measured data

In this example, the measured data fall within a smooth distribution.

```python
measured_data = pandas.read_csv("Parsed data/All Cause Mortality.csv")
measured_data.replace({0: numpy.nan}, inplace=True) # remove zeros, which are redacted data, not measured 0
seaborn.histplot(measured_data["Percent Deaths 25+"])
```
